```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import math
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
score = pd.read_csv("question_counts.csv", encoding='latin1')
score=score[24000:36000]
```

```python
indices1=score.query('java>3 or python>3 or c>3').index
indices2=score.query('sql>2 or plsql>2 or oracle>2').index
indices3=score.query('php>3 or angularjs>3 or javascript>3 or html>3 or css>3').index
```

```python
score_array=score.as_matrix(columns=score.columns[1:])
```

```python
print("begun")
score_embedded = TSNE(n_components=2,random_state=None).fit_transform(score_array)
print("done")
```

```python
target_score_array1=[]
for i in indices1:
    temp_array=[score_embedded[i-24000][0],score_embedded[i-24000][1]]
    target_score_array1.append(temp_array)
    temp_array=[]
    
target_score_array1=np.array(target_score_array1)
```

```python
target_score_array2=[]
for i in indices2:
    temp_array=[score_embedded[i-24000][0],score_embedded[i-24000][1]]
    target_score_array2.append(temp_array)
    temp_array=[]
    
target_score_array2=np.array(target_score_array2)
```

```python
target_score_array3=[]
for i in indices3:
    temp_array=[score_embedded[i-24000][0],score_embedded[i-24000][1]]
    target_score_array3.append(temp_array)
    temp_array=[]
    
target_score_array3=np.array(target_score_array3)
```

```python
x=score_embedded[:,0]
y=score_embedded[:,1]
fig = plt.figure(figsize = (10,10), dpi=100)
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Component 1', fontsize = 10)
ax.set_ylabel('Component 2', fontsize = 10)
ax.set_title('Plot for TSNE', fontsize = 10)
ax.scatter(x, y, c='purple')
plt.show()
```

```python
x=score_embedded[:,0]
y=score_embedded[:,1]
fig = plt.figure(figsize = (10,10), dpi=100)
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Component 1', fontsize = 10)
ax.set_ylabel('Component 2', fontsize = 10)
ax.set_title('Plot for TSNE with some back-end technology results', fontsize = 10)
ax.scatter(x, y, c='purple')
ax.scatter(target_score_array1[:,0],target_score_array1[:,1],color="red")
plt.show()
```

```python
x=score_embedded[:,0]
y=score_embedded[:,1]
fig = plt.figure(figsize = (5,5), dpi=100)
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Component 1', fontsize = 10)
ax.set_ylabel('Component 2', fontsize = 10)
ax.set_title('Plot for TSNE with some back-end and database technology results', fontsize = 8)
ax.scatter(x, y, c='purple')
ax.scatter(target_score_array1[:,0],target_score_array1[:,1],color="red")
ax.scatter(target_score_array2[:,0],target_score_array2[:,1],color="cyan")

plt.show()
```

```python
x=score_embedded[:,0]
y=score_embedded[:,1]
fig = plt.figure(figsize = (5,5), dpi=100)
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Component 1', fontsize = 10)
ax.set_ylabel('Component 2', fontsize = 10)
ax.set_title('Plot for TSNE with some back-end, front-end and database technology results', fontsize = 8)
ax.scatter(x, y, c="purple")
ax.scatter(target_score_array1[:,0],target_score_array1[:,1],color="red")
ax.scatter(target_score_array2[:,0],target_score_array2[:,1],color="cyan")

ax.scatter(target_score_array3[:,0],target_score_array3[:,1],color="yellow")

plt.show()
```
